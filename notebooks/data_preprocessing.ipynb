```python
import pandas as pd
import holidays
import plotly.express as px
from utilsforecast.plotting import plot_series
from datetime import timedelta
pd.set_option('display.max_columns', 20)
pd.set_option('display.width', 200)
```

# 2022

```python
df_2022 =  pd.read_excel("/mnt/data/oe215/rhindrikson/datasets/load/entsoe/tables/monthly_hourly_load_values_2022.xlsx")
# filter for germany
df_de_2022 = df_2022[df_2022["CountryCode"] == "DE"]
df_de_2022 = df_de_2022.drop(columns=["TimeTo", "TimeFrom", "CreateDate", "MeasureItem", "DateShort", "CountryCode", "Cov_ratio", "Value_ScaleTo100"])
df_de_2022['unique_id'] = "load"
df_de_2022.rename(columns={"DateUTC": "ds", "Value": "y"}, inplace=True) 
df_de_2022["week_day"] = df_de_2022["ds"].dt.dayofweek
# Hodilays
de_holidays = holidays.Germany(years=2022)
df_de_2022["is_holiday"] = df_de_2022["ds"].dt.date.isin(de_holidays).astype(int)
df_de_2022['day_before_holiday'] = (df_de_2022['ds'] + pd.Timedelta(days=1)).dt.date.isin(de_holidays).astype(int)
df_de_2022['day_after_holiday'] = (df_de_2022['ds'] - pd.Timedelta(days=1)).dt.date.isin(de_holidays).astype(int)
```

# Future exogenous variables

```python
# 1. First, determine the last date in your training data
last_date = df_combined['ds'].max()
print(last_date)

# 2. Create future dates for the horizon
future_dates = pd.date_range(
    start=last_date + timedelta(hours=1),
    periods=8766,
    freq='h'
)

futr_df = pd.DataFrame({
    "ds": future_dates,
    "unique_id": "load"
})

# 4. Add week_day
futr_df["week_day"] = futr_df["ds"].dt.dayofweek

# 5. Create holiday calendar for BOTH 2025 and 2026
de_holidays = holidays.Germany(years=[2025, 2026])

# 6. Add holiday features
futr_df["is_holiday"] = futr_df["ds"].dt.date.isin(de_holidays).astype(int)
futr_df['day_before_holiday'] = (futr_df['ds'] + pd.Timedelta(days=1)).dt.date.isin(de_holidays).astype(int)
futr_df['day_after_holiday'] = (futr_df['ds'] - pd.Timedelta(days=1)).dt.date.isin(de_holidays).astype(int)

print(futr_df.head())
print(futr_df.tail())
print(f"Date range: {futr_df['ds'].min()} to {futr_df['ds'].max()}")
```
